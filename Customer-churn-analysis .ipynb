```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv("Telco-Customer-Churn.csv")

# Show first rows
print(df.head())

# Dataset info
print(df.info())

# Convert TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Remove missing values
df = df.dropna()

# -------------------------------
# 1. Churn Distribution
# -------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()

# -------------------------------
# 2. Churn by Contract Type
# -------------------------------
plt.figure(figsize=(7,5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.show()

# -------------------------------
# 3. Churn by Internet Service
# -------------------------------
plt.figure(figsize=(7,5))
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Churn by Internet Service")
plt.show()

# -------------------------------
# 4. Monthly Charges vs Churn
# -------------------------------
plt.figure(figsize=(7,5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()

# -------------------------------
# 5. Tenure vs Churn
# -------------------------------
plt.figure(figsize=(7,5))
sns.histplot(data=df, x="tenure", hue="Churn", bins=30)
plt.title("Tenure Distribution vs Churn")
plt.show()

# -------------------------------
# 6. Churn Rate
# -------------------------------
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print("Churn Rate:")
print(churn_rate)

# -------------------------------
# 7. Customer Lifetime Value
# -------------------------------
df['CLV'] = df['MonthlyCharges'] * df['tenure']

avg_clv = df['CLV'].mean()
print("Average Customer Lifetime Value:", avg_clv)

# -------------------------------
# 8. Retention by Contract
# -------------------------------
retention_contract = pd.crosstab(df['Contract'], df['Churn'])

retention_contract.plot(kind='bar', stacked=True, figsize=(7,5))
plt.title("Customer Retention by Contract Type")
plt.ylabel("Number of Customers")
plt.show()
```
